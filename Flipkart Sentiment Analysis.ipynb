```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
reviews = pd.read_csv('data-analytics/archive/electronics_reviews_uniq.csv')
reviews.head()
```

```python
reviews = reviews.drop('reviewed_at',axis=1)
```

```python
reviews[reviews.isna().any(axis=1)]
```

```python
reviews = reviews.dropna(axis=1)
```

```python
honor = reviews[['product','rating','likes','dislikes','description']][reviews['product']=='Honor 10 Lite (Midnight Black, 32 GB)']
honor.head()
```

```python
# description into lower case
honor['description'] = honor.description.astype(str).apply(lambda x: x.lower())
honor.head()
```

```python
#  Removed punctuation from description
import re, string
honor['description'] = honor['description'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' , x))
honor.head()
```

```python
# Removed numbers from description
def rmnum(s):
    t = ''
    for i in s:
        if not i.isdigit():
            t+=i
    return t
honor['description'] = honor.description.apply(rmnum)
honor.head()
```

```python
# Removed stopwords using nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words(['english']))
stop_words.add('honor')
stop_words.add('midnight')
stop_words.add('black')
stop_words.add('lite')

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
honor['description'] = honor['description'].apply(lambda x: remove_stopwords(x))
honor.head()
```

```python
#  Reduced a word to its word stem using lemmatization

nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def stem_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
honor["description"] = honor["description"].apply(lambda x: stem_words(x))
honor.head()
```

```python
# pos tagged description
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
from nltk import word_tokenize

def postag(x):
    text_tok = nltk.word_tokenize(x)
    pos_tagged = nltk.pos_tag(text_tok)
    return dict(pos_tagged)
honor["description"] = honor["description"].apply(lambda x: postag(x))
honor.head()
```
